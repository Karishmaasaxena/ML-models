```python
import cv2

def draw_bounding_box(image , boxes , labels):
  # Check if image loading was successful before proceeding
  if image is None:
    print("Error: Could not load image. Please check the image path.")
    return  # Exit the function if image is None

  image_height, image_width, _ = image.shape
  print(image_width)
  for box , label in zip(boxes , labels):
    x_min , y_min, x_max , y_max = boxes
    x_min = int(x_min * image_width )
    y_min = int(y_min * image_height )
    x_max = int(x_max * image_width )
    y_max = int(y_max * image_height)
    cv2.rectangle(image , (x_min , y_min) , (x_max , y_max ), (255,0,0),2)
    cv2.putText(image , str(label) , (x_min , y_min -10), cv2.FONT_HERSHEY_SIMPLEX, .9 , (255,0,0,),2)


image = cv2.imread('/content/cycle.jpg')

# Check if the image was loaded successfully
if image is not None:
  boxes = [30, 20 ,400, 300]
  labels = [0]

  draw_bounding_box(image , boxes , labels)

  cv2_imshow(image)
else:
  print("Error: Image loading failed.")
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('yolov8n.pt')


cap = cv2.VideoCapture("/content/zina.mp4")


while True:
    ret, frame = cap.read()

    results = model(frame, stream = True)

    for result in results:
        boxes = result.boxes.xyxy
        confidences = result.boxes.conf
        class_ids = result.boxes.cls

        for box , confidence , class_id in zip(boxes , confidences, class_ids):
            x1, y1 , x2, y2 = map(int, box)
            label = model.names[int(class_id)]
            confidence = float(confidence)


            cv2.rectangle(frame, (x1, y1), (x2,y2), (0,255,0), 2)


            label_text = f'{label}: {confidence}'

            cv2.putText(frame, label_text, (x1,y1-10), cv2.FONT_HERSHEY_PLAIN, .5 , (0,255,0),2)


    cv2.imshow('YOLO',frame)


cap.release()

```

```python
import cv2
from ultralytics import YOLO
from google.colab.patches import cv2_imshow # Import cv2_imshow

model = YOLO('yolov8n.pt')

cap = cv2.VideoCapture("/content/zina.mp4")

while True:
    ret, frame = cap.read()

    results = model(frame, stream = True)

    for result in results:
        boxes = result.boxes.xyxy
        confidences = result.boxes.conf
        class_ids = result.boxes.cls

        for box , confidence , class_id in zip(boxes , confidences, class_ids):
            x1, y1 , x2, y2 = map(int, box)
            label = model.names[int(class_id)]
            confidence = float(confidence)

            cv2.rectangle(frame, (x1, y1), (x2,y2), (0,255,0), 2)

            label_text = f'{label}: {confidence}'

            cv2.putText(frame, label_text, (x1,y1-10), cv2.FONT_HERSHEY_PLAIN, .5 , (0,255,0),2)

    cv2_imshow(frame) # Use cv2_imshow instead of cv2.imshow

cap.release()
```
